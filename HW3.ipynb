```python
class ImplicitEuler(solver):
    def __init__(self, nx=100, CFL=0.5, a_=1, init=None, bcs=None):
        super().__init__(nx, CFL, a_, init, bcs)
        self._list=[]
        self.matrix=np.zeros((nx+1, nx+1))
        self.matrix[0][-1]=-CFL/2
        self.matrix[-1][0]=CFL/2
        self.matrix[0][1]=CFL/2
        self.matrix[-1][-2]=-CFL/2
        self.matrix+=np.eye(nx+1)
        for i in range(1,len(self.matrix)-1):
            self.matrix[i][i-1]=-CFL/2
            self.matrix[i][i+1]=CFL/2
        print(self.matrix)
        #self.matrix=np.linalg.inv(self.matrix)
        #print(self.matrix)
        print('Method: Implicit Euler 1st order, nx: {}, dx: {}, CFL: {}, dt: {}'.format(self.nx, self.dx, self.CFL, self.dt))
    
    def method(self, u_curr, u_next, tt):
        #return self.CFL*(np.roll(u_curr,shift=1)-np.roll(u_curr,shift=-1))+u_curr, u_next
        #return self.matrix@u_curr, u_next
        return np.linalg.solve(self.matrix,u_curr), u_next
    
    def solve(self, nt):
        return solver.solve(self,nt,arg_plot=None,arg_save=set_save(self._list))
    
    def plot_result2(self, _range):
        print(len(self._list))
        plt.figure(dpi=120)
        set_axis(plt.gca(),ymaj=1,ymin=0.25)
        for i in _range:
            plt.plot(self.x_range,self._list[i][1],label=r'$t=$'+'{:.4f}'.format(self._list[i][0]*self.dt)+r', $i=$'+'{:d}'.format(self._list[i][0]))
            #plt.text(0.25,1.75,'t={:.5f}'.format(FTCS_list1[i][0]*testFTCS_1.dt))
        plt.ylim(-0.2, 3.2)
        plt.legend()#loc='upper left'
        plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设置
a = 1.0
Mx = 100
CFL = 0.5
x_left, x_right = -0.5, 0.5
dx = (x_right - x_left) / Mx
dt = CFL * dx / a  # CFL = a*dt/dx → dt = CFL*dx/a

# 空间网格
x = np.linspace(x_left + 0.5*dx, x_right - 0.5*dx, Mx)  # 网格点中心坐标

# 初始条件
u0 = np.where((x >= -0.25) & (x < 0.25), 1.0, 0.0)

# 构造系数矩阵A
C = a * dt / dx
A = np.zeros((Mx, Mx))
for j in range(Mx):
    A[j, j] = 1 + C  # 主对角线元素
    if j > 0:
        A[j, j-1] = -C  # 下对角线元素
    else:
        A[j, Mx-1] = -C  # 周期性边界条件处理

print(A)
# 需要计算的时间点
target_times = [0.0, 0.1, 1.0, 10.0]

# 存储解和对应时间
solutions = []
current_u = u0.copy()
current_time = 0.0

# 初始时间t=0
solutions.append((current_time, current_u.copy()))

# 按时间顺序计算每个目标时间点
sorted_times = sorted(target_times)
for t_target in sorted_times[1:]:  # 跳过0.0
    delta_t = t_target - current_time
    n_steps = int(round(delta_t / dt))
    # 时间步进
    for _ in range(n_steps):
        current_u = np.linalg.solve(A, current_u)
        current_time += dt
    solutions.append((current_time, current_u.copy()))

# 绘制结果
plt.figure(figsize=(10, 6))
for t, u in solutions:
    plt.plot(x, u, label=f't={t:.1f}')

plt.title('Implicit Euler Solution at Different Times')
plt.xlabel('x')
plt.ylabel('u')
plt.legend()
plt.grid(True)
plt.show()

# 分析间断附近的行为（以t=0.1和t=1.0为例）
# 观察解的耗散和波形展宽情况
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设置
a = 1.0
Mx = 100
CFL = 0.5
x_left, x_right = -0.5, 0.5
dx = (x_right - x_left) / Mx
dt = CFL * dx / a  # 保持CFL条件一致
C = a * dt / dx  # 计算Courant数

# 空间网格
x = np.linspace(x_left + 0.5*dx, x_right - 0.5*dx, Mx)  # 网格中心坐标

# 初始条件
u0 = np.where((x >= -0.25) & (x < 0.25), 1.0, 0.0)

# 构造系数矩阵A（三对角循环矩阵）
A = np.zeros((Mx, Mx))
for j in range(Mx):
    # 主对角线
    A[j, j] = 1.0
    # 下对角线 (j-1项)
    left = j-1 if j > 0 else Mx-1
    A[j, left] = -C/2
    # 上对角线 (j+1项)
    right = j+1 if j < Mx-1 else 0
    A[j, right] = C/2

# 需要计算的时间点
target_times = [0.0, 0.1, 1.0, 10.0]

# 存储解和对应时间
solutions = []
current_u = u0.copy()
current_time = 0.0

# 初始时间t=0
solutions.append((current_time, current_u.copy()))

# 按时间顺序计算每个目标时间点
sorted_times = sorted(target_times)
for t_target in sorted_times[1:]:  # 跳过0.0
    delta_t = t_target - current_time
    n_steps = int(round(delta_t / dt))
    # 时间步进
    for _ in range(n_steps):
        current_u = np.linalg.solve(A, current_u)
        current_time += dt
    solutions.append((current_time, current_u.copy()))

# 绘图
plt.figure(figsize=(10, 6))
for t, u in solutions:
    plt.plot(x, u, label=f't={t:.1f}')

plt.title('Central Difference Implicit Scheme Solution')
plt.xlabel('x')
plt.ylabel('u')
plt.legend()
plt.grid(True)
plt.show()
```
